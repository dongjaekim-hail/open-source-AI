## Python 과 Jupyter

파이썬 (python) 은 원래 인터프리터 언어입니다. 인터프리터 언어란, 대화형이라는 그 말 뜻 그대로 한 번에 한 줄의 코드를 실행하고 그 코드의 실행 결과를 바로 확인하는 형태의 코드를 의미합니다.

아마도 여러분들은 이런 인터프리터의 형태로 파이썬을 활용하는 일이 그다지 많지 않습니다. 디버깅을 제외하고는 .py 확장자의 파일을 눌러서 실행하죠! 이런 파일들은 다른 절차형 프로그래밍 언어처럼 순서대로 실행이 됩니다.

상호작용을 통해 (interative) 파이썬을 인터프리터 형태로 활용할 수 있도록 만들어 진 것이 ipython 이라고 하는 쉘 (shell) 입니다. 여기서는 실행하는 코드의 line number 가 직혀나오기 때문에 어떤 코드를 실행했을 때 어떤 아웃풋이 나오는지를 직관적으로 확인할 수 있습니다. 그러나 이것은 파이썬에 제한적이기 때문에, 새로운 형태가 제안됩니다.

그게 바로 jupyter입니다.

파이썬이 주로 쓰이는 분야중에 하나가 데이터사이언스 입니다. 이 분야에서 사용하는 다른 유명 언어로는 R이 있구요, matlab을 대체하는 또 다른 언어로 Julia가 있습니다.

(R 은 통계 분석에 정확하고 다양한 기능이 있다는 장점이 있구요, julia는 machine learning 류의 분석이 빠르다는 장점이 있습니다.)

julia + python + R 을 합쳐서 jupyter라고 부르는 새로운 인터랙티브 대화형 쉘로 만들어진 것이 jupyter 이고 colab 은 그 문법을 그대로 활용합니다!

## Colab이란?

colab 은 클라우드 컴퓨팅으로 만들어진 jupyter notebook 입니다. 저장공간을 google drive 와 연결할 수 있기 때문에 큰 용량의 데이터를 처리하기에도 적절합니다.

### 대충 어떤 식으로 구현되나?

어러분이 colab을 열면 해당 jupyter notebook 파일이 클라우드 (즉, 서버) 에서 열립니다. 서버에서 virtual box와 같은 어떤 프로그램(도커)을 통해서 가상의 리눅스 컴퓨터를 여러분의 jupyter notebook 에 연결 시켜줍니다.

여러분이 이 노트북에서 활용하고 하는 모든 연산은 클라우드에 만들어진 가상의 리눅스 컴퓨터를 활용하여 이루어집니다.

기본적으로 colab에는 여러분들이 주로 활용하는 프로그램들이 설치되어있습니다. 대표적인 것들이 python 의 library 들입니다. pytorch, tensorflow 같은 것들은 당연하고, opencv등 유용하게 활용하는 library들이 미리 설치되어 있습니다.

어떤 library가 설치되어있는지는 `!pip list` 를 실행하면 알 수 있습니다. 이는 jupyter notebook 실행을 통해서 실제로 연결되어있는 리눅스 컴퓨터의 terminal 에 명령어를 전달하는 방법입니다. `!pwd` 라고 하면 현재의 working directory 를 알려주게 되겠죠.
